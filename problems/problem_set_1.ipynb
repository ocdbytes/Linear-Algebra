```python
A = np.array([[0.8, 0.3], [0.2, 0.7]])
u = np.array([[1], [0]])
v = np.array([[0], [1]])
```

```python
# Calculating u1...u7

x = np.copy(u)
k = np.arange(8)

# Iterate and update x
while x.shape[1] <= 7:
    u = A @ u  # Matrix multiplication
    x = np.hstack((x, u))  # Append new u as a column

# Plot the results
plt.plot(k, x.T)  # Transpose x to align with k
plt.xlabel("k")
plt.ylabel("Values")
plt.legend(["x1", "x2"])
plt.grid()
plt.show()
```

```python
# Calculating v1...v7

x = np.copy(v)
k = np.arange(8)

# Iterate and update x
while x.shape[1] <= 7:
    v = A @ v  # Matrix multiplication
    x = np.hstack((x, v))  # Append new u as a column

# Plot the results
plt.plot(k, x.T)  # Transpose x to align with k
plt.xlabel("k")
plt.ylabel("Values")
plt.legend(["x1", "x2"])
plt.grid()
plt.show()

print("Thus steady state vector is :")
print(np.array([[0.6], [0.4]]))
```
