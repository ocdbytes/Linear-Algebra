```python
# Create the matrix K
K = 2 * np.eye(9) + np.diag(-1 * np.ones(8), 1) + np.diag(-1 * np.ones(8), -1)

# Create the vector b
b = 10 * np.ones((9, 1))

# Solve the linear system Kx = b
x = np.linalg.solve(K, b)

# Plot the values in x
plt.plot(range(1, 10), x, marker='o', linestyle='-', color='b')
plt.title('Solution of Kx = b')
plt.xlabel('Index')
plt.ylabel('x values')
plt.grid(True)
plt.show()
```
